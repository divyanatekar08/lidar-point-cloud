```python
import trimesh
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the PLY file
mesh = trimesh.load("cloudcompare-exercise-7.ply")
points = mesh.vertices
df = pd.DataFrame(points, columns=["X", "Y", "Z"])

# Try to load colors if available
if hasattr(mesh, 'visual') and hasattr(mesh.visual, 'vertex_colors'):
    colors = mesh.visual.vertex_colors[:, :3] / 255.0  # Normalize RGB values
else:
    colors = 'blue'  # fallback color

# Plot the point cloud
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Slightly reduced point size
ax.scatter(df["X"], df["Y"], df["Z"], c=colors, s=5, alpha=0.8)

ax.set_title("3D Point Cloud (Smaller Size & RGB if Available)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.set_xlim(df["X"].min(), df["X"].max())
ax.set_ylim(df["Y"].min(), df["Y"].max())
ax.set_zlim(df["Z"].min(), df["Z"].max())

plt.tight_layout()
plt.show()
```
